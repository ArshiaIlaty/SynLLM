```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
import ast
import re
```

```python
# Set up matplotlib and seaborn
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100

# Load the evaluation results
results_df = pd.read_csv('/home/jovyan/SynLLM/opensource/Cirrhosis/Eval/cirrhosis_evaluation_summary.csv')

# Display basic information
print(f"Total models evaluated: {results_df['model'].nunique()}")
print(f"Total prompts evaluated: {results_df['prompt'].nunique()}")
print(f"Total model × prompt combinations: {len(results_df)}")
print("\nModels evaluated:")
for model in sorted(results_df['model'].unique()):
    print(f"- {model}")

# Clean up model names for better display
results_df['model_display'] = results_df['model'].apply(lambda x: x.split('_')[-1] if '_' in x else x)
```

```python
# Function to make column names more readable
def clean_column_name(col):
    return col.replace('_', ' ').title()

# Basic data exploration
print("\nColumns in the dataset:")
metrics_categories = {}
for col in results_df.columns:
    if '_' in col and col not in ['model', 'prompt', 'model_display']:
        category = col.split('_')[0]
        if category not in metrics_categories:
            metrics_categories[category] = []
        metrics_categories[category].append(col)

for category, metrics in metrics_categories.items():
    print(f"\n{clean_column_name(category)} ({len(metrics)} metrics)")
    for metric in metrics[:4]:  # Show just a few examples
        print(f"  - {metric}")
    if len(metrics) > 4:
        print(f"  - ... and {len(metrics) - 4} more")

# 1. Distribution similarity (lower is better)
wasserstein_cols = [col for col in results_df.columns if 'distribution_metrics' in col and 'wasserstein' in col]
print(f"Available Wasserstein columns: {wasserstein_cols}")
results_df['distribution_score'] = results_df[wasserstein_cols].mean(axis=1)

# 2. Statistical test score (lower is better)
ks_cols = [col for col in results_df.columns if 'numerical_metrics' in col and 'ks_statistic' in col]
print(f"Available KS statistic columns: {ks_cols}")
results_df['stat_test_score'] = results_df[ks_cols].mean(axis=1)

# 3. Medical consistency score (higher ratio is better - closer to real data)
# Check if the expected medical consistency columns exist
if 'medical_consistency_glucose_stroke_relationship_real_difference' in results_df.columns and \
   'medical_consistency_glucose_stroke_relationship_synthetic_difference' in results_df.columns:
    results_df['medical_consistency_score'] = np.abs(
        results_df['medical_consistency_glucose_stroke_relationship_synthetic_difference'] / 
        results_df['medical_consistency_glucose_stroke_relationship_real_difference']
    )
else:
    # Use any available medical consistency columns
    med_cols = [col for col in results_df.columns if 'medical_consistency' in col and 'synthetic_difference' in col]
    if med_cols:
        real_col = med_cols[0].replace('synthetic_difference', 'real_difference')
        if real_col in results_df.columns:
            results_df['medical_consistency_score'] = np.abs(results_df[med_cols[0]] / results_df[real_col])
        else:
            results_df['medical_consistency_score'] = 1.0  # Default value
    else:
        results_df['medical_consistency_score'] = 1.0  # Default value

# 4. Privacy score (higher is better)
results_df['privacy_score'] = results_df['privacy_assessment_nn_distance_ratio']

# 5. Correlation matrix similarity (lower distance is better)
results_df['correlation_score'] = results_df['feature_correlations_correlation_matrix_distance']

# 6. Categorical distributions (lower chi2 is better)
chi2_cols = [col for col in results_df.columns if 'categorical_similarity' in col and 'chi2_statistic' in col]
print(f"Available Chi-square columns: {chi2_cols}")
if chi2_cols:
    results_df['categorical_score'] = results_df[chi2_cols].mean(axis=1)
else:
    results_df['categorical_score'] = 0.0  # Default value

# Create ranking function
def rank_models(df, score_column, ascending=True):
    """Rank models by a score column. Default is ascending (lower is better)"""
    return df.sort_values(score_column, ascending=ascending)[[
        'model_display', 'prompt', score_column
    ]].reset_index(drop=True)
```

```python
# ------------------- Visualizations -------------------

# Helper function to safely get numeric columns
def get_numeric_columns(df, col_list):
    numeric_cols = []
    for col in col_list:
        if col in df.columns:
            # Try to convert to numeric, dropping any rows with non-numeric values
            try:
                # Check if column has any non-numeric values
                pd.to_numeric(df[col], errors='raise')
                numeric_cols.append(col)
            except:
                # Skip columns with non-numeric values
                print(f"Skipping column {col} due to non-numeric values")
                continue
    return numeric_cols

# 1. Overall performance by model
plt.figure(figsize=(12, 10))
metrics = ['distribution_score', 'stat_test_score', 'correlation_score', 'categorical_score']

# Make sure all metrics exist in the dataframe and are numeric
metrics = get_numeric_columns(results_df, metrics)

# Add optional metrics if they exist and are numeric
optional_metrics = ['medical_consistency_score', 'privacy_score']
for metric in optional_metrics:
    if metric in results_df.columns:
        try:
            pd.to_numeric(results_df[metric], errors='raise')
            metrics.append(metric)
        except:
            print(f"Skipping metric {metric} due to non-numeric values")

# Normalize metrics to 0-1 scale for comparison
normalized_df = pd.DataFrame()
for metric in metrics:
    if metric in ['medical_consistency_score', 'privacy_score'] and metric in results_df.columns:
        # Higher is better - invert
        normalized_df[metric] = 1 - (
            (pd.to_numeric(results_df[metric], errors='coerce') - 
             pd.to_numeric(results_df[metric], errors='coerce').min()) / 
            (pd.to_numeric(results_df[metric], errors='coerce').max() - 
             pd.to_numeric(results_df[metric], errors='coerce').min() + 1e-10)
        )
    else:
        # Lower is better
        normalized_df[metric] = (
            (pd.to_numeric(results_df[metric], errors='coerce') - 
             pd.to_numeric(results_df[metric], errors='coerce').min()) / 
            (pd.to_numeric(results_df[metric], errors='coerce').max() - 
             pd.to_numeric(results_df[metric], errors='coerce').min() + 1e-10)
        )
normalized_df['model_display'] = results_df['model_display']
normalized_df['prompt'] = results_df['prompt']

# Calculate mean normalized score across all metrics
normalized_df['overall_score'] = normalized_df[metrics].mean(axis=1)

# Plot overall model performance
plt.figure(figsize=(14, 8))
model_means = normalized_df.groupby('model_display')[metrics + ['overall_score']].mean().reset_index()
sns.barplot(x='model_display', y='overall_score', data=model_means)
plt.title('Overall Model Performance (Lower is Better)', fontsize=16)
plt.xlabel('Model', fontsize=14)
plt.ylabel('Normalized Score (0-1)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('overall_model_performance.png')

# 2. Model performance by prompt
plt.figure(figsize=(14, 10))
pivot_df = normalized_df.pivot_table(
    index='model_display', 
    columns='prompt', 
    values='overall_score',
    aggfunc='mean'
)
sns.heatmap(pivot_df, annot=True, cmap='viridis_r', fmt='.2f')
plt.title('Model Performance by Prompt (Lower is Better)', fontsize=16)
plt.tight_layout()
plt.savefig('model_performance_by_prompt.png')

# 3. Distribution similarity - using numerical statistics
plt.figure(figsize=(14, 8))

# Search for appropriate numerical metrics for key features
key_features = ['Age', 'Bilirubin', 'Albumin', 'Cholesterol']
dist_metrics = []

# Look for mean difference metrics first
for feature in key_features:
    col_name = f'numerical_statistics_{feature}_mean_difference'
    if col_name in results_df.columns:
        try:
            pd.to_numeric(results_df[col_name], errors='raise')
            dist_metrics.append(col_name)
        except:
            print(f"Skipping {col_name} due to non-numeric values")

# If no mean difference metrics found, look in advanced metrics
if len(dist_metrics) == 0:
    for feature in key_features:
        col_name = f'advanced_metrics_numerical_metrics_{feature}'
        if col_name in results_df.columns:
            try:
                pd.to_numeric(results_df[col_name], errors='raise')
                dist_metrics.append(col_name)
            except:
                print(f"Skipping {col_name} due to non-numeric values")

# If some metrics were found, create the distribution similarity plot
if len(dist_metrics) > 0:
    dist_df = results_df.groupby('model_display')[dist_metrics].mean().reset_index()
    dist_df_melted = pd.melt(dist_df, id_vars=['model_display'], value_vars=dist_metrics,
                            var_name='Feature', value_name='Difference')
    
    # Clean up feature names
    if 'numerical_statistics' in dist_metrics[0]:
        dist_df_melted['Feature'] = dist_df_melted['Feature'].apply(
            lambda x: x.replace('numerical_statistics_', '').replace('_mean_difference', '')
        )
    else:
        dist_df_melted['Feature'] = dist_df_melted['Feature'].apply(
            lambda x: x.replace('advanced_metrics_numerical_metrics_', '')
        )
    
    sns.barplot(x='model_display', y='Difference', hue='Feature', data=dist_df_melted)
    plt.title('Distribution Similarity by Feature (Lower is Better)', fontsize=16)
    plt.xlabel('Model', fontsize=14)
    plt.ylabel('Difference', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Feature')
    plt.tight_layout()
    plt.savefig('distribution_similarity.png')
else:
    print("No suitable distribution metrics found for plotting")

# 4. Categorical variable similarity using Chi-square statistics
plt.figure(figsize=(14, 8))
chi2_columns = [col for col in results_df.columns if 'chi2_statistic' in col]

# Filter to only include numeric columns
numeric_chi2_columns = []
for col in chi2_columns:
    try:
        pd.to_numeric(results_df[col], errors='raise')
        numeric_chi2_columns.append(col)
    except:
        print(f"Skipping {col} due to non-numeric values")

if numeric_chi2_columns:
    med_df = results_df.groupby('model_display')[numeric_chi2_columns].mean().reset_index()
    
    # For categorical similarities
    plt.figure(figsize=(12, 6))
    med_df_melted = pd.melt(med_df, id_vars=['model_display'], value_vars=numeric_chi2_columns,
                            var_name='Feature', value_name='Chi-Square')
    med_df_melted['Feature'] = med_df_melted['Feature'].apply(
        lambda x: x.replace('categorical_similarity_', '').replace('_chi2_statistic', '')
    )
    sns.barplot(x='model_display', y='Chi-Square', hue='Feature', data=med_df_melted)
    plt.title('Categorical Variable Similarity (Lower Chi-Square is Better)', fontsize=16)
    plt.xlabel('Model', fontsize=14)
    plt.ylabel('Chi-Square Statistic', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Feature')
    plt.tight_layout()
    plt.savefig('categorical_similarity.png')
else:
    print("No suitable chi-square metrics found for plotting")

# 5. Correlation matrix similarity
if 'correlation_score' in metrics:
    plt.figure(figsize=(12, 6))
    # Convert to numeric with coerce to handle any string values
    results_df['correlation_score_numeric'] = pd.to_numeric(results_df['correlation_score'], errors='coerce')
    corr_df = results_df.groupby('model_display')['correlation_score_numeric'].mean().reset_index()
    sns.barplot(x='model_display', y='correlation_score_numeric', data=corr_df)
    plt.title('Correlation Matrix Similarity (Lower is Better)', fontsize=16)
    plt.xlabel('Model', fontsize=14)
    plt.ylabel('Frobenius Distance', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('correlation_similarity.png')

# 6. Privacy metrics
privacy_metric = 'privacy_assessment_nn_distance_ratio'
if privacy_metric in results_df.columns:
    try:
        # Convert to numeric with coerce to handle any string values
        results_df['privacy_metric_numeric'] = pd.to_numeric(results_df[privacy_metric], errors='coerce')
        
        plt.figure(figsize=(12, 6))
        privacy_df = results_df.groupby('model_display')['privacy_metric_numeric'].mean().reset_index()
        sns.barplot(x='model_display', y='privacy_metric_numeric', data=privacy_df)
        plt.axhline(y=1.0, color='r', linestyle='--', label='Threshold (>1 is better)')
        plt.title('Privacy Protection (Higher is Better)', fontsize=16)
        plt.xlabel('Model', fontsize=14)
        plt.ylabel('Nearest Neighbor Distance Ratio', fontsize=14)
        plt.xticks(rotation=45, ha='right')
        plt.legend()
        plt.tight_layout()
        plt.savefig('privacy_protection.png')
    except Exception as e:
        print(f"Error plotting privacy metrics: {e}")

# Find the best performing models - safely handling potential string values
best_models = {}

# Helper function to rank models safely
def safe_rank_models(df, column, ascending=True):
    try:
        # Convert to numeric with coerce to handle any string values
        df[f'{column}_numeric'] = pd.to_numeric(df[column], errors='coerce')
        return df.sort_values(f'{column}_numeric', ascending=ascending).dropna(subset=[f'{column}_numeric'])
    except Exception as e:
        print(f"Error ranking by {column}: {e}")
        return pd.DataFrame()  # Return empty DataFrame on error

# Best for distribution similarity
if 'distribution_score' in metrics:
    ranked = safe_rank_models(results_df, 'distribution_score', ascending=True)
    if not ranked.empty:
        best_models['Distribution Similarity'] = ranked.iloc[0]

# Best for statistical tests
if 'stat_test_score' in metrics:
    ranked = safe_rank_models(results_df, 'stat_test_score', ascending=True)
    if not ranked.empty:
        best_models['Statistical Tests'] = ranked.iloc[0]

# Best for medical consistency
if 'medical_consistency_score' in metrics:
    ranked = safe_rank_models(results_df, 'medical_consistency_score', ascending=False)
    if not ranked.empty:
        best_models['Medical Consistency'] = ranked.iloc[0]

# Best for privacy
if 'privacy_score' in metrics:
    ranked = safe_rank_models(results_df, 'privacy_score', ascending=False)
    if not ranked.empty:
        best_models['Privacy Protection'] = ranked.iloc[0]

# Best for correlation matrix
if 'correlation_score' in metrics:
    ranked = safe_rank_models(results_df, 'correlation_score', ascending=True)
    if not ranked.empty:
        best_models['Correlation Matrix Similarity'] = ranked.iloc[0]

# Best for categorical distributions
if 'categorical_score' in metrics:
    ranked = safe_rank_models(results_df, 'categorical_score', ascending=True)
    if not ranked.empty:
        best_models['Categorical Distributions'] = ranked.iloc[0]

# Best overall (lowest average normalized score)
best_models['Overall Performance'] = normalized_df.iloc[normalized_df['overall_score'].idxmin()]

# Create summary DataFrame of best models
if best_models:
    best_models_df = pd.DataFrame(best_models).T.reset_index()
    print("Columns in best_models_df:", best_models_df.columns.tolist())
    print("Shape of best_models_df:", best_models_df.shape)

    # Display best models
    print("\n=== Best Performing Models by Category ===")
    display(best_models_df)

# Create rankings for all models
model_rankings = normalized_df.groupby('model_display')['overall_score'].mean().sort_values().reset_index()
model_rankings.columns = ['Model', 'Average Score']
model_rankings['Rank'] = model_rankings.index + 1

# Display rankings
print("\n=== Overall Model Rankings ===")
display(model_rankings)
```

```python
# ------------------- Detailed Analysis of Top Models -------------------

# Get the top 3 performing models
top_models = model_rankings.head(3)['Model'].tolist()

# Extract detailed metrics for these models
top_models_df = results_df[results_df['model_display'].isin(top_models)]

# Analyze the best performing prompts for top models
top_prompts = normalized_df[
    normalized_df['model_display'].isin(top_models)
].groupby(['model_display', 'prompt'])['overall_score'].mean().reset_index()
top_prompts = top_prompts.sort_values('overall_score')

print("\n=== Best Prompt for Each Top Model ===")
for model in top_models:
    best_prompt = top_prompts[top_prompts['model_display'] == model].iloc[0]['prompt']
    score = top_prompts[top_prompts['model_display'] == model].iloc[0]['overall_score']
    print(f"{model}: {best_prompt} (Score: {score:.4f})")

# Create summary visualization of top models across metrics
plt.figure(figsize=(14, 10))
top_models_metrics = results_df[results_df['model_display'].isin(top_models)].groupby('model_display')[
    ['distribution_score', 'stat_test_score', 'medical_consistency_score', 
     'privacy_score', 'correlation_score', 'categorical_score']
].mean().reset_index()

# Normalize metrics for radar chart
normalized_metrics = top_models_metrics.copy()
for col in top_models_metrics.columns[1:]:
    if col in ['medical_consistency_score', 'privacy_score']:
        # Higher is better - invert for visualization
        min_val = top_models_metrics[col].min()
        max_val = top_models_metrics[col].max()
        normalized_metrics[col] = 1 - ((top_models_metrics[col] - min_val) / (max_val - min_val + 1e-10))
    else:
        # Lower is better
        min_val = top_models_metrics[col].min()
        max_val = top_models_metrics[col].max()
        normalized_metrics[col] = (top_models_metrics[col] - min_val) / (max_val - min_val + 1e-10)

# Prepare data for radar chart
categories = ['Distribution', 'Statistical Tests', 'Medical Consistency', 
              'Privacy', 'Correlation', 'Categorical']
N = len(categories)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # Close the loop

# Create radar chart
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

# Draw one axis per variable and add labels
plt.xticks(angles[:-1], categories, fontsize=12)

# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([0.25, 0.5, 0.75], ["0.25", "0.5", "0.75"], fontsize=10)
plt.ylim(0, 1)

# Plot each model
for i, model in enumerate(normalized_metrics['model_display']):
    values = normalized_metrics.iloc[i, 1:].values.tolist()
    values += values[:1]  # Close the loop
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=model)
    ax.fill(angles, values, alpha=0.1)

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.title('Top Models Performance Comparison (Lower is Better)', fontsize=15)
plt.tight_layout()
plt.savefig('top_models_radar.png')

# Summary statistics of the entire dataset
print("\n=== Summary Statistics of All Metrics ===")
metrics_summary = results_df[[
    'distribution_score', 'stat_test_score', 'medical_consistency_score',
    'privacy_score', 'correlation_score', 'categorical_score'
]].describe()
display(metrics_summary)

# Conclusions
print("""
=== Analysis Conclusions ===

1. Overall Performance: The models with the best overall performance across all metrics are:
   {}

2. Key Findings:
   - Distribution Similarity: Models vary significantly in their ability to match the real data distributions
   - Medical Consistency: Top models preserve the relationship between medical variables
   - Privacy Protection: Some models provide better privacy guarantees than others
   - Correlation Preservation: The best models maintain similar correlation structures to real data

3. Best Prompts:
   Different prompts work best for different models, but in general, prompt {} tends to produce
   the most realistic synthetic data across multiple models.

4. Recommendations:
   - For general use: {} with prompt {}
   - For maximum privacy: {} with prompt {}
   - For best medical consistency: {} with prompt {}
""".format(
    ', '.join(top_models),
    top_prompts.groupby('prompt')['overall_score'].mean().idxmin(),
    model_rankings.iloc[0]['Model'], top_prompts[top_prompts['model_display'] == model_rankings.iloc[0]['Model']].iloc[0]['prompt'],
    best_models['Privacy Protection']['model_display'], best_models['Privacy Protection']['prompt'],
    best_models['Medical Consistency']['model_display'], best_models['Medical Consistency']['prompt']
))
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.ticker as ticker

# Set plotting style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")
plt.rcParams['figure.figsize'] = [14, 8]
plt.rcParams['figure.dpi'] = 100

# Load computational performance data
with open('/home/jovyan/SynLLM/opensource/Diabetes/reports/usage.jsonl', 'r') as f:
    usage_data = [json.loads(line) for line in f]

usage_df = pd.DataFrame(usage_data)

# Clean model names for better display
usage_df['model_display'] = usage_df['model'].apply(lambda x: x.split('/')[-1] if '/' in x else x)

# Load quality metrics if available (assuming results_df exists from previous analysis)
# If not, comment this part out or load the evaluation results
try:
    # Check if results_df exists from previous cells
    existing_df = results_df
    has_quality_metrics = True
except NameError:
    try:
        # Try to load it from CSV
        results_df = pd.read_csv('evaluation_summary.csv')  
        has_quality_metrics = True
    except:
        print("Quality metrics not available. Only showing computational metrics.")
        has_quality_metrics = False

# 1. Computational Performance Visualizations

# 1.1 Generation Time by Model and Prompt
plt.figure(figsize=(16, 8))
sns.barplot(x='model_display', y='duration_sec', hue='prompt', data=usage_df)
plt.title('Generation Time by Model and Prompt', fontsize=18)
plt.xlabel('Model', fontsize=14)
plt.ylabel('Duration (seconds)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Prompt Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add text labels on bars
for i, prompt in enumerate(usage_df['prompt'].unique()):
    prompt_data = usage_df[usage_df['prompt'] == prompt]
    for j, (_, row) in enumerate(prompt_data.iterrows()):
        plt.text(j + (i - 1.5) * 0.25, row['duration_sec'] + 10, 
                 f"{row['duration_sec']:.0f}s", 
                 ha='center', va='bottom', fontsize=8, rotation=90)

plt.tight_layout()
plt.savefig('generation_time_comparison.png')
plt.show()

# 1.2 Resource Usage Comparison
plt.figure(figsize=(16, 10))

# Create a subplot for CPU usage
plt.subplot(2, 1, 1)
cpu_data = usage_df.groupby('model_display')['cpu_percent'].mean().sort_values(ascending=False).reset_index()
sns.barplot(x='model_display', y='cpu_percent', hue='model_display', data=cpu_data, palette='Blues_d', legend=False)
plt.title('Average CPU Usage by Model', fontsize=16)
plt.xlabel('')
plt.ylabel('CPU Usage (%)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Create a subplot for GPU memory usage
plt.subplot(2, 1, 2)
usage_df['gpu_mem_start_MB'] = pd.to_numeric(usage_df['gpu_mem_start_MB'])
gpu_data = usage_df.groupby('model_display')['gpu_mem_start_MB'].mean().sort_values(ascending=False).reset_index()
sns.barplot(x='model_display', y='gpu_mem_start_MB', hue='model_display', data=gpu_data, palette='Oranges_d', legend=False)
plt.title('Average GPU Memory Usage by Model', fontsize=16)
plt.xlabel('Model', fontsize=14)
plt.ylabel('GPU Memory (MB)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('resource_usage_comparison.png')
plt.show()

# 1.3 Generation Time by Prompt Type (all models)
plt.figure(figsize=(12, 7))
prompt_time = usage_df.groupby('prompt')['duration_sec'].mean().reset_index()
sns.barplot(x='prompt', y='duration_sec', hue='prompt', data=prompt_time, palette='viridis', legend=False)
plt.title('Average Generation Time by Prompt Type', fontsize=18)
plt.xlabel('Prompt Type', fontsize=14)
plt.ylabel('Average Duration (seconds)', fontsize=14)
for i, v in enumerate(prompt_time['duration_sec']):
    plt.text(i, v + 5, f"{v:.1f}s", ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('prompt_time_comparison.png')
plt.show()

# 2. Combined Quality and Computational Analysis (if quality metrics are available)
if has_quality_metrics:
    # Merge performance and quality data
    # Assuming 'model' and 'prompt' are the keys to join on
    results_df['model_display'] = results_df['model'].apply(lambda x: x.split('_')[-1] if '_' in x else x)
    
    # Create a merged dataframe with both metrics
    combined_df = pd.merge(
        usage_df, 
        results_df, 
        on=['model_display', 'prompt'], 
        how='inner',
        suffixes=('_usage', '_quality')
    )
    
    # If the merge doesn't have enough data, try an alternative approach
    if len(combined_df) < 5:
        print("Warning: Not enough matching data between usage and quality metrics.")
        print("Using alternative approach to combine metrics.")
        
        # Aggregate usage data by model_display and prompt
        usage_agg = usage_df.groupby(['model_display', 'prompt']).agg({
            'duration_sec': 'mean',
            'cpu_percent': 'mean',
            'gpu_mem_start_MB': lambda x: pd.to_numeric(x, errors='coerce').mean()
        }).reset_index()
        
        # Aggregate quality data by model_display and prompt
        quality_agg = results_df.groupby(['model_display', 'prompt']).agg({
            'distribution_score': 'mean' if 'distribution_score' in results_df.columns else 'first',
            'medical_consistency_score': 'mean' if 'medical_consistency_score' in results_df.columns else 'first',
            'overall_score': 'mean' if 'overall_score' in results_df.columns else 'first'
        }).reset_index()
        
        # Merge aggregated data
        combined_df = pd.merge(
            usage_agg, 
            quality_agg, 
            on=['model_display', 'prompt'], 
            how='inner'
        )
    
    if len(combined_df) > 0:
        # 2.1 Efficiency Plot (Quality vs Speed)
        plt.figure(figsize=(14, 10))
        
        # Create scatter plot with model names as annotations
        # Lower score is better for quality, lower time is better for speed
        if 'overall_score' in combined_df.columns:
            quality_metric = 'overall_score'
        elif 'distribution_score' in combined_df.columns:
            quality_metric = 'distribution_score'
        else:
            # Use the first numeric column that might represent quality
            numeric_cols = combined_df.select_dtypes(include=['number']).columns
            quality_cols = [col for col in numeric_cols if 'score' in col.lower()]
            quality_metric = quality_cols[0] if quality_cols else numeric_cols[0]
        
        plt.scatter(
            combined_df['duration_sec'], 
            combined_df[quality_metric],
            s=combined_df['gpu_mem_start_MB'] / 100,  # Size based on GPU memory
            alpha=0.7,
            c=combined_df.index,  # Color based on index for variety
            cmap='viridis'
        )
        
        # Add model names as annotations
        for i, row in combined_df.iterrows():
            plt.annotate(
                f"{row['model_display']}\n{row['prompt']}", 
                (row['duration_sec'], row[quality_metric]),
                fontsize=8,
                xytext=(5, 5),
                textcoords='offset points'
            )
        
        plt.title('Efficiency Analysis: Quality vs. Speed', fontsize=18)
        plt.xlabel('Generation Time (seconds)', fontsize=14)
        plt.ylabel(f'Quality Score ({quality_metric})', fontsize=14)
        plt.grid(True, linestyle='--', alpha=0.7)
        
        # Add quadrant lines and labels
        x_mid = combined_df['duration_sec'].median()
        y_mid = combined_df[quality_metric].median()
        
        plt.axvline(x=x_mid, color='gray', linestyle='--', alpha=0.5)
        plt.axhline(y=y_mid, color='gray', linestyle='--', alpha=0.5)
        
        plt.text(combined_df['duration_sec'].min(), y_mid, "Faster", ha='left', va='bottom')
        plt.text(combined_df['duration_sec'].max(), y_mid, "Slower", ha='right', va='bottom')
        plt.text(x_mid, combined_df[quality_metric].min(), "Better Quality", ha='right', va='bottom')
        plt.text(x_mid, combined_df[quality_metric].max(), "Lower Quality", ha='right', va='top')
        
        plt.tight_layout()
        plt.savefig('efficiency_analysis.png')
        plt.show()
        
        # 2.2 Trade-off Matrix (Heatmap)
        plt.figure(figsize=(16, 12))
        
        # Create pivot tables for different metrics
        pivot_time = combined_df.pivot_table(
            index='model_display', 
            columns='prompt', 
            values='duration_sec',
            aggfunc='mean'
        )
        
        pivot_quality = combined_df.pivot_table(
            index='model_display', 
            columns='prompt', 
            values=quality_metric,
            aggfunc='mean'
        )
        
        # Normalize values for color mapping
        norm_time = (pivot_time - pivot_time.min().min()) / (pivot_time.max().max() - pivot_time.min().min())
        norm_quality = (pivot_quality - pivot_quality.min().min()) / (pivot_quality.max().max() - pivot_quality.min().min())
        
        # Create a composite score (low is good for both time and quality)
        composite_score = (norm_time + norm_quality) / 2
        
        # Create a custom color map (green=good, red=bad)
        cmap = LinearSegmentedColormap.from_list("custom_cmap", ["green", "yellow", "red"])
        
        # Plot the heatmap
        plt.subplot(1, 1, 1)
        ax = sns.heatmap(
            composite_score, 
            annot=True, 
            cmap=cmap, 
            fmt='.2f',
            linewidths=0.5
        )
        plt.title('Performance-Quality Trade-off Matrix (Lower is Better)', fontsize=18)
        plt.tight_layout()
        plt.savefig('tradeoff_matrix.png')
        plt.show()
        
        # 2.3 Top Models Comparison
        # Select top 5 models based on composite score
        top_models = composite_score.mean(axis=1).sort_values().index[:5]
        
        # Filter data for top models
        top_models_data = combined_df[combined_df['model_display'].isin(top_models)]
        
        # Create a radar chart for multiple dimensions
        plt.figure(figsize=(12, 10))
        
        # Define metrics to compare (normalize all so lower is better)
        metrics = [
            ('duration_sec', 'Generation Time'),
            (quality_metric, 'Quality Score'),
            ('cpu_percent', 'CPU Usage'),
            ('gpu_mem_start_MB', 'GPU Memory')
        ]
        
        # Number of metrics
        N = len(metrics)
        
        # Create angles for radar chart
        angles = [n / float(N) * 2 * np.pi for n in range(N)]
        angles += angles[:1]  # Close the loop
        
        # Initialize the radar chart
        ax = plt.subplot(111, polar=True)
        
        # Draw one axis per variable and add labels
        plt.xticks(angles[:-1], [label for _, label in metrics])
        
        # Draw y-labels
        ax.set_rlabel_position(0)
        plt.yticks([0.25, 0.5, 0.75], ["0.25", "0.5", "0.75"], color="grey", size=7)
        plt.ylim(0, 1)
        
        # Plot each model
        for model in top_models:
            model_data = combined_df[combined_df['model_display'] == model]
            if len(model_data) == 0:
                continue
                
            # Get average values for the model
            values = []
            for metric, _ in metrics:
                # Normalize metric values (min-max scaling)
                min_val = combined_df[metric].min()
                max_val = combined_df[metric].max()
                avg_val = model_data[metric].mean()
                
                # Scale so lower values are better
                norm_val = (avg_val - min_val) / (max_val - min_val) if max_val > min_val else 0
                values.append(norm_val)
            
            # Close the loop
            values += values[:1]
            
            # Plot the model
            ax.plot(angles, values, linewidth=1, linestyle='solid', label=model)
            ax.fill(angles, values, alpha=0.1)
        
        # Add legend
        plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
        plt.title('Top Models Performance Comparison (Lower is Better)', size=15)
        plt.tight_layout()
        plt.savefig('top_models_radar.png')
        plt.show()
        
        # 2.4 Efficiency Ranking Table
        # Create a ranking based on efficiency (balancing quality and speed)
        ranking_df = combined_df.groupby('model_display', group_keys=False).apply(
            lambda x: pd.Series({
                'avg_duration': x['duration_sec'].mean(),
                'avg_quality': x[quality_metric].mean(),
                'efficiency_score': (
                    (x['duration_sec'] - combined_df['duration_sec'].min()) / 
                    (combined_df['duration_sec'].max() - combined_df['duration_sec'].min()) +
                    (x[quality_metric] - combined_df[quality_metric].min()) / 
                    (combined_df[quality_metric].max() - combined_df[quality_metric].min())
                ).mean() / 2
            }), include_groups=False
        ).reset_index()
        
        # Sort by efficiency score (lower is better)
        ranking_df = ranking_df.sort_values('efficiency_score')
        ranking_df['rank'] = range(1, len(ranking_df) + 1)
        
        # Display the ranking table
        print("\n=== Model Efficiency Ranking (Quality vs. Speed) ===")
        print(ranking_df[['rank', 'model_display', 'avg_duration', 'avg_quality', 'efficiency_score']])
        
        # Save the ranking table as a figure
        plt.figure(figsize=(12, len(ranking_df) * 0.4 + 1))
        ax = plt.subplot(111, frame_on=False)
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
        
        table = plt.table(
            cellText=ranking_df[['rank', 'model_display', 'avg_duration', 'avg_quality', 'efficiency_score']].values,
            colLabels=['Rank', 'Model', 'Avg Duration (s)', 'Avg Quality', 'Efficiency Score'],
            cellLoc='center',
            loc='center',
            bbox=[0, 0, 1, 1]
        )
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        table.scale(1, 1.5)
        
        # Highlight top 3 models
        for i in range(3):
            cells = [table[i+1, j] for j in range(5)]
            for cell in cells:
                cell.set_facecolor('#D5F5E3')  # Light green
        
        plt.title('Model Efficiency Ranking (Lower Score is Better)', fontsize=16, y=0.99)
        plt.tight_layout()
        plt.savefig('efficiency_ranking.png')
        plt.show()
    else:
        print("Could not match quality and usage data. Check if model and prompt names match between datasets.")
```
