```python
# Function to make column names more readable
def clean_column_name(col):
    return col.replace('_', ' ').title()

# Basic data exploration
print("\nColumns in the dataset:")
metrics_categories = {}
for col in results_df.columns:
    if '_' in col and col not in ['model', 'prompt', 'model_display']:
        category = col.split('_')[0]
        if category not in metrics_categories:
            metrics_categories[category] = []
        metrics_categories[category].append(col)

for category, metrics in metrics_categories.items():
    print(f"\n{clean_column_name(category)} ({len(metrics)} metrics)")
    for metric in metrics[:4]:  # Show just a few examples
        print(f"  - {metric}")
    if len(metrics) > 4:
        print(f"  - ... and {len(metrics) - 4} more")

# 1. Distribution similarity (lower is better)
wasserstein_cols = [col for col in results_df.columns if 'distribution_metrics' in col and 'wasserstein' in col]
print(f"Available Wasserstein columns: {wasserstein_cols}")
results_df['distribution_score'] = results_df[wasserstein_cols].mean(axis=1)

# 2. Statistical test score (lower is better)
ks_cols = [col for col in results_df.columns if 'numerical_metrics' in col and 'ks_statistic' in col]
print(f"Available KS statistic columns: {ks_cols}")
results_df['stat_test_score'] = results_df[ks_cols].mean(axis=1)

# 3. Medical consistency score (higher ratio is better - closer to real data)
# Check if the expected medical consistency columns exist
if 'medical_consistency_glucose_stroke_relationship_real_difference' in results_df.columns and \
   'medical_consistency_glucose_stroke_relationship_synthetic_difference' in results_df.columns:
    results_df['medical_consistency_score'] = np.abs(
        results_df['medical_consistency_glucose_stroke_relationship_synthetic_difference'] / 
        results_df['medical_consistency_glucose_stroke_relationship_real_difference']
    )
else:
    # Use any available medical consistency columns
    med_cols = [col for col in results_df.columns if 'medical_consistency' in col and 'synthetic_difference' in col]
    if med_cols:
        real_col = med_cols[0].replace('synthetic_difference', 'real_difference')
        if real_col in results_df.columns:
            results_df['medical_consistency_score'] = np.abs(results_df[med_cols[0]] / results_df[real_col])
        else:
            results_df['medical_consistency_score'] = 1.0  # Default value
    else:
        results_df['medical_consistency_score'] = 1.0  # Default value

# 4. Privacy score (higher is better)
# List all key privacy metrics you want to include
privacy_metrics = [
    'privacy_assessment_age_gender_violations_violation_rate',  # Lower is better
    'advanced_metrics_privacy_metrics_nn_distance_ratio',       # Higher is better
    'advanced_metrics_privacy_metrics_identifiability_score',   # Lower is better
    'privacy_assessment_overall_privacy_score'                  # Higher is better
]

# Filter to only those present in the DataFrame
privacy_metrics_available = [col for col in privacy_metrics if col in results_df.columns]

# Normalize each metric so that higher is always better
for col in privacy_metrics_available:
    if 'violation_rate' in col or 'identifiability_score' in col:
        # Lower is better, so invert
        max_val = results_df[col].max()
        min_val = results_df[col].min()
        results_df[f'norm_{col}'] = 1 - (results_df[col] - min_val) / (max_val - min_val + 1e-8)
    else:
        # Higher is better, normalize directly
        max_val = results_df[col].max()
        min_val = results_df[col].min()
        results_df[f'norm_{col}'] = (results_df[col] - min_val) / (max_val - min_val + 1e-8)

# Aggregate normalized privacy metrics into a composite privacy score
norm_cols = [f'norm_{col}' for col in privacy_metrics_available]
results_df['privacy_score'] = results_df[norm_cols].mean(axis=1)


# 5. Correlation matrix similarity (lower distance is better)
results_df['correlation_score'] = results_df['feature_correlations_correlation_matrix_distance']

# 6. Categorical distributions (lower chi2 is better)
chi2_cols = [col for col in results_df.columns if 'categorical_similarity' in col and 'chi2_statistic' in col]
print(f"Available Chi-square columns: {chi2_cols}")
if chi2_cols:
    results_df['categorical_score'] = results_df[chi2_cols].mean(axis=1)
else:
    results_df['categorical_score'] = 0.0  # Default value

# Create ranking function
def rank_models(df, score_column, ascending=True):
    """Rank models by a score column. Default is ascending (lower is better)"""
    return df.sort_values(score_column, ascending=ascending)[[
        'model_display', 'prompt', score_column
    ]].reset_index(drop=True)
```

```python
# ------------------- Visualizations -------------------

# Helper function to safely get numeric columns
def get_numeric_columns(df, col_list):
    numeric_cols = []
    for col in col_list:
        if col in df.columns:
            # Try to convert to numeric, dropping any rows with non-numeric values
            try:
                # Check if column has any non-numeric values
                pd.to_numeric(df[col], errors='raise')
                numeric_cols.append(col)
            except:
                # Skip columns with non-numeric values
                print(f"Skipping column {col} due to non-numeric values")
                continue
    return numeric_cols

# 1. Overall performance by model
plt.figure(figsize=(12, 10))
metrics = ['distribution_score', 'stat_test_score', 'correlation_score', 'categorical_score']

# Make sure all metrics exist in the dataframe and are numeric
metrics = get_numeric_columns(results_df, metrics)

# Add optional metrics if they exist and are numeric
optional_metrics = ['medical_consistency_score', 'privacy_score']
for metric in optional_metrics:
    if metric in results_df.columns:
        try:
            pd.to_numeric(results_df[metric], errors='raise')
            metrics.append(metric)
        except:
            print(f"Skipping metric {metric} due to non-numeric values")

# Normalize metrics to 0-1 scale for comparison
normalized_df = pd.DataFrame()
for metric in metrics:
    if metric in ['medical_consistency_score', 'privacy_score'] and metric in results_df.columns:
        # Higher is better - invert
        normalized_df[metric] = 1 - (
            (pd.to_numeric(results_df[metric], errors='coerce') - 
             pd.to_numeric(results_df[metric], errors='coerce').min()) / 
            (pd.to_numeric(results_df[metric], errors='coerce').max() - 
             pd.to_numeric(results_df[metric], errors='coerce').min() + 1e-10)
        )
    else:
        # Lower is better
        normalized_df[metric] = (
            (pd.to_numeric(results_df[metric], errors='coerce') - 
             pd.to_numeric(results_df[metric], errors='coerce').min()) / 
            (pd.to_numeric(results_df[metric], errors='coerce').max() - 
             pd.to_numeric(results_df[metric], errors='coerce').min() + 1e-10)
        )
normalized_df['model_display'] = results_df['model_display']
normalized_df['prompt'] = results_df['prompt']

# Calculate mean normalized score across all metrics
normalized_df['overall_score'] = normalized_df[metrics].mean(axis=1)

# Plot overall model performance
plt.figure(figsize=(14, 8))
model_means = normalized_df.groupby('model_display')[metrics + ['overall_score']].mean().reset_index()
sns.barplot(x='model_display', y='overall_score', data=model_means)
plt.title('Overall Model Performance (Lower is Better)', fontsize=16)
plt.xlabel('Model', fontsize=14)
plt.ylabel('Normalized Score (0-1)', fontsize=14)
plt.xticks(rotation=45, ha='right')



# 2. Model performance by prompt
plt.figure(figsize=(10, 10))
pivot_df = normalized_df.pivot_table(
    index='model_display', 
    columns='prompt', 
    values='overall_score',
    aggfunc='mean'
)

# Create the heatmap and store the colorbar object
heatmap = sns.heatmap(pivot_df, annot=True, cmap='viridis_r', fmt='.2f',
            annot_kws={'size': 14, 'weight': 'bold'},
            cbar_kws={
                'label': 'Score',
                'shrink': 0.8     # Optional: adjust colorbar size
            })

# Access and modify the colorbar
cbar = heatmap.collections[0].colorbar
# Set label with larger font and bold
cbar.ax.set_ylabel('Score', fontsize=16, fontweight='bold')
# Increase tick font size
cbar.ax.tick_params(labelsize=14)

# Increase tick label size for x and y axes
plt.xticks(fontsize=14)
plt.yticks(fontsize=14, fontweight='bold')  # Make model names bold

plt.title('Model Performance by Prompt (Lower is Better)', 
          fontsize=20,
          pad=20,
          weight='bold')
plt.xlabel('Prompt', fontsize=16, weight='bold')
plt.ylabel('Model', fontsize=16, weight='bold')
# plt.tight_layout()
save_all_formats(plt.gcf(), 'model_performance_by_prompt')

# 3. Distribution similarity - using numerical statistics
plt.figure(figsize=(14, 8))

# Search for appropriate numerical metrics for key features
key_features = ['Age', 'Bilirubin', 'Albumin', 'Cholesterol']
dist_metrics = []

# Look for mean difference metrics first
for feature in key_features:
    col_name = f'numerical_statistics_{feature}_mean_difference'
    if col_name in results_df.columns:
        try:
            pd.to_numeric(results_df[col_name], errors='raise')
            dist_metrics.append(col_name)
        except:
            print(f"Skipping {col_name} due to non-numeric values")

# If no mean difference metrics found, look in advanced metrics
if len(dist_metrics) == 0:
    for feature in key_features:
        col_name = f'advanced_metrics_numerical_metrics_{feature}'
        if col_name in results_df.columns:
            try:
                pd.to_numeric(results_df[col_name], errors='raise')
                dist_metrics.append(col_name)
            except:
                print(f"Skipping {col_name} due to non-numeric values")

# If some metrics were found, create the distribution similarity plot
if len(dist_metrics) > 0:
    dist_df = results_df.groupby('model_display')[dist_metrics].mean().reset_index()
    dist_df_melted = pd.melt(dist_df, id_vars=['model_display'], value_vars=dist_metrics,
                            var_name='Feature', value_name='Difference')
    
    # Clean up feature names
    if 'numerical_statistics' in dist_metrics[0]:
        dist_df_melted['Feature'] = dist_df_melted['Feature'].apply(
            lambda x: x.replace('numerical_statistics_', '').replace('_mean_difference', '')
        )
    else:
        dist_df_melted['Feature'] = dist_df_melted['Feature'].apply(
            lambda x: x.replace('advanced_metrics_numerical_metrics_', '')
        )
    
    sns.barplot(x='model_display', y='Difference', hue='Feature', data=dist_df_melted)
    plt.title('Distribution Similarity by Feature (Lower is Better)', fontsize=16)
    plt.xlabel('Model', fontsize=14)
    plt.ylabel('Difference', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Feature')
    plt.tight_layout()
    plt.savefig('distribution_similarity.png')
else:
    print("No suitable distribution metrics found for plotting")

# 4. Categorical variable similarity using Chi-square statistics
plt.figure(figsize=(14, 8))
chi2_columns = [col for col in results_df.columns if 'chi2_statistic' in col]

# Filter to only include numeric columns
numeric_chi2_columns = []
for col in chi2_columns:
    try:
        pd.to_numeric(results_df[col], errors='raise')
        numeric_chi2_columns.append(col)
    except:
        print(f"Skipping {col} due to non-numeric values")

if numeric_chi2_columns:
    med_df = results_df.groupby('model_display')[numeric_chi2_columns].mean().reset_index()
    
    # For categorical similarities
    plt.figure(figsize=(12, 6))
    med_df_melted = pd.melt(med_df, id_vars=['model_display'], value_vars=numeric_chi2_columns,
                            var_name='Feature', value_name='Chi-Square')
    med_df_melted['Feature'] = med_df_melted['Feature'].apply(
        lambda x: x.replace('categorical_similarity_', '').replace('_chi2_statistic', '')
    )
    sns.barplot(x='model_display', y='Chi-Square', hue='Feature', data=med_df_melted)
    plt.title('Categorical Variable Similarity (Lower Chi-Square is Better)', fontsize=16)
    plt.xlabel('Model', fontsize=14)
    plt.ylabel('Chi-Square Statistic', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Feature')
    plt.tight_layout()
    plt.savefig('categorical_similarity.png')
else:
    print("No suitable chi-square metrics found for plotting")

# 5. Correlation matrix similarity
if 'correlation_score' in metrics:
    plt.figure(figsize=(12, 6))
    # Convert to numeric with coerce to handle any string values
    results_df['correlation_score_numeric'] = pd.to_numeric(results_df['correlation_score'], errors='coerce')
    corr_df = results_df.groupby('model_display')['correlation_score_numeric'].mean().reset_index()
    sns.barplot(x='model_display', y='correlation_score_numeric', data=corr_df)
    plt.title('Correlation Matrix Similarity (Lower is Better)', fontsize=16)
    plt.xlabel('Model', fontsize=14)
    plt.ylabel('Frobenius Distance', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig('correlation_similarity.png')

# 6. Privacy metrics
privacy_metric = 'privacy_assessment_nn_distance_ratio'
if privacy_metric in results_df.columns:
    try:
        # Convert to numeric with coerce to handle any string values
        results_df['privacy_metric_numeric'] = pd.to_numeric(results_df[privacy_metric], errors='coerce')
        
        plt.figure(figsize=(12, 6))
        privacy_df = results_df.groupby('model_display')['privacy_metric_numeric'].mean().reset_index()
        sns.barplot(x='model_display', y='privacy_metric_numeric', data=privacy_df)
        plt.axhline(y=1.0, color='r', linestyle='--', label='Threshold (>1 is better)')
        plt.title('Privacy Protection (Higher is Better)', fontsize=16)
        plt.xlabel('Model', fontsize=14)
        plt.ylabel('Nearest Neighbor Distance Ratio', fontsize=14)
        plt.xticks(rotation=45, ha='right')
        plt.legend()
        plt.tight_layout()
        plt.savefig('privacy_protection.png')
    except Exception as e:
        print(f"Error plotting privacy metrics: {e}")

# Find the best performing models - safely handling potential string values
best_models = {}

# Helper function to rank models safely
def safe_rank_models(df, column, ascending=True):
    try:
        # Convert to numeric with coerce to handle any string values
        df[f'{column}_numeric'] = pd.to_numeric(df[column], errors='coerce')
        return df.sort_values(f'{column}_numeric', ascending=ascending).dropna(subset=[f'{column}_numeric'])
    except Exception as e:
        print(f"Error ranking by {column}: {e}")
        return pd.DataFrame()  # Return empty DataFrame on error

# Best for distribution similarity
if 'distribution_score' in metrics:
    ranked = safe_rank_models(results_df, 'distribution_score', ascending=True)
    if not ranked.empty:
        best_models['Distribution Similarity'] = ranked.iloc[0]

# Best for statistical tests
if 'stat_test_score' in metrics:
    ranked = safe_rank_models(results_df, 'stat_test_score', ascending=True)
    if not ranked.empty:
        best_models['Statistical Tests'] = ranked.iloc[0]

# Best for medical consistency
if 'medical_consistency_score' in metrics:
    ranked = safe_rank_models(results_df, 'medical_consistency_score', ascending=False)
    if not ranked.empty:
        best_models['Medical Consistency'] = ranked.iloc[0]

# Best for privacy
if 'privacy_score' in metrics:
    ranked = safe_rank_models(results_df, 'privacy_score', ascending=False)
    if not ranked.empty:
        best_models['Privacy Protection'] = ranked.iloc[0]

# Best for correlation matrix
if 'correlation_score' in metrics:
    ranked = safe_rank_models(results_df, 'correlation_score', ascending=True)
    if not ranked.empty:
        best_models['Correlation Matrix Similarity'] = ranked.iloc[0]

# Best for categorical distributions
if 'categorical_score' in metrics:
    ranked = safe_rank_models(results_df, 'categorical_score', ascending=True)
    if not ranked.empty:
        best_models['Categorical Distributions'] = ranked.iloc[0]

# Best overall (lowest average normalized score)
best_models['Overall Performance'] = normalized_df.iloc[normalized_df['overall_score'].idxmin()]

# Create summary DataFrame of best models
if best_models:
    best_models_df = pd.DataFrame(best_models).T.reset_index()
    print("Columns in best_models_df:", best_models_df.columns.tolist())
    print("Shape of best_models_df:", best_models_df.shape)

    # Display best models
    print("\n=== Best Performing Models by Category ===")
    display(best_models_df)

# Create rankings for all models
model_rankings = normalized_df.groupby('model_display')['overall_score'].mean().sort_values().reset_index()
model_rankings.columns = ['Model', 'Average Score']
model_rankings['Rank'] = model_rankings.index + 1

# Display rankings
print("\n=== Overall Model Rankings ===")
display(model_rankings)
```

```python
# 2. Combined Quality and Computational Analysis (if quality metrics are available)
if has_quality_metrics:
    # Merge performance and quality data
    display_names = {
    "gpt2": "GPT-2-Small",
    "openai-community_gpt2-large": "GPT-2-Large",
    "openai-community_gpt2-medium": "GPT-2-Medium",
    "mistralai_Mistral-7B-Instruct-v0.2": "Mistral 7B",
    "meta-llama_Llama-2-7b-chat-hf": "LLaMA 2 7B",
    "meta-llama_Llama-3.1-8b-Instruct": "LLaMA 3.1 8B",
    "google_gemma-7b-it": "Gemma 7B",
    "01-ai_Yi-6B-Chat": "Yi 6B",
    "HuggingFaceH4_zephyr-7b-beta": "Zephyr 7B",
    "Qwen_Qwen2-7B-Instruct": "Qwen2 7B",
    "internlm_internlm2_5-7b-chat": "InternLM2.5 7B",
    "stabilityai_StableBeluga-7B": "StableBeluga 7B",
    "openchat_openchat-3.5-0106": "OpenChat 3.5",
    "NousResearch_Nous-Hermes-2-Yi-34B": "Nous Hermes Yi 34B",
    "mosaicml_mpt-7b-instruct": "Mosaic 7B",
    "openchat_openchat_3.5": "OpenChat 3.5",
    "TheBloke_openchat_3.5-GPTQ": "OpenChat 3.5 GPTQ",
    "meta-llama/Llama-2-13b-hf" : "Llama 2 13B HF",
    "meta-llama/Llama-2-13b-chat" : "Llama 2 13B Chat",
    "meta-llama/Llama-2-13b-chat-hf" : "Llama 2 13B Chat HF",
    "meta-llama/Meta-Llama-3-8B" : "Meta Llama 3 8B",
    "NousResearch/Nous-Hermes-2-Mistral-7B-DPO" : "Nous Hermes Mistral 7B DPO",
    }
    
    results_df['model_display'] = results_df['model'].map(display_names).fillna(results_df['model'])

    
    # Create a merged dataframe with both metrics
    combined_df = pd.merge(
        usage_df, 
        results_df, 
        on=['model_display', 'prompt'], 
        how='inner',
        suffixes=('_usage', '_quality')
    )
    
    # If the merge doesn't have enough data, try an alternative approach
    if len(combined_df) < 5:
        print("Warning: Not enough matching data between usage and quality metrics.")
        print("Using alternative approach to combine metrics.")
        
        # Aggregate usage data by model_display and prompt
        usage_agg = usage_df.groupby(['model_display', 'prompt']).agg({
            'duration_sec': 'mean',
            'cpu_percent': 'mean',
            'gpu_mem_start_MB': lambda x: pd.to_numeric(x, errors='coerce').mean()
        }).reset_index()
        
        # Aggregate quality data by model_display and prompt
        quality_agg = results_df.groupby(['model_display', 'prompt']).agg({
            'distribution_score': 'mean',
            'medical_consistency_score': 'mean',
            'privacy_score': 'mean',
            'correlation_score': 'mean',
            'categorical_score': 'mean',
            'stat_test_score': 'mean'
        }).reset_index()
        
        # Calculate overall score
        score_columns = [col for col in quality_agg.columns if col.endswith('_score')]
        quality_agg['overall_score'] = quality_agg[score_columns].mean(axis=1)
        
        # Merge with usage data if available
        if not usage_df.empty and 'model_display' in usage_df.columns:
            combined_df = pd.merge(
                usage_agg,
                quality_agg,
                on=['model_display', 'prompt'],
                how='inner'
            )
        else:
            print("Warning: No usage data available. Proceeding with quality metrics only.")
            combined_df = quality_agg.copy()
        
        print(f"Shape of combined dataframe: {combined_df.shape}")
    
    if len(combined_df) > 0:
        # 2.1 Efficiency Plot (Quality vs Speed)
        plt.figure(figsize=(14, 10))
        
        # Create scatter plot with model names as annotations
        # Lower score is better for quality, lower time is better for speed
        if 'overall_score' in combined_df.columns:
            quality_metric = 'overall_score'
        elif 'distribution_score' in combined_df.columns:
            quality_metric = 'distribution_score'
        else:
            # Use the first numeric column that might represent quality
            numeric_cols = combined_df.select_dtypes(include=['number']).columns
            quality_cols = [col for col in numeric_cols if 'score' in col.lower()]
            quality_metric = quality_cols[0] if quality_cols else numeric_cols[0]
        
        plt.scatter(
            combined_df['duration_sec'], 
            combined_df[quality_metric],
            s=combined_df['gpu_mem_start_MB'] / 100,  # Size based on GPU memory
            alpha=0.7,
            c=combined_df.index,  # Color based on index for variety
            cmap='viridis'
        )
        
        # Add model names as annotations
        for i, row in combined_df.iterrows():
            plt.annotate(
                f"{row['model_display']}\n{row['prompt']}", 
                (row['duration_sec'], row[quality_metric]),
                fontsize=8,
                xytext=(5, 5),
                textcoords='offset points'
            )
        
        plt.title('Efficiency Analysis: Quality vs. Speed', fontsize=18)
        plt.xlabel('Generation Time (seconds)', fontsize=14)
        plt.ylabel(f'Quality Score ({quality_metric})', fontsize=14)
        plt.grid(True, linestyle='--', alpha=0.7)
        
        # Add quadrant lines and labels
        x_mid = combined_df['duration_sec'].median()
        y_mid = combined_df[quality_metric].median()
        
        plt.axvline(x=x_mid, color='gray', linestyle='--', alpha=0.5)
        plt.axhline(y=y_mid, color='gray', linestyle='--', alpha=0.5)
        
        plt.text(combined_df['duration_sec'].min(), y_mid, "Faster", ha='left', va='bottom')
        plt.text(combined_df['duration_sec'].max(), y_mid, "Slower", ha='right', va='bottom')
        plt.text(x_mid, combined_df[quality_metric].min(), "Better Quality", ha='right', va='bottom')
        plt.text(x_mid, combined_df[quality_metric].max(), "Lower Quality", ha='right', va='top')
        
        plt.tight_layout()
        save_all_formats(plt.gcf(), 'efficiency_analysis')
        plt.close()
        
        # 2.2 Trade-off Matrix (Heatmap)
        plt.figure(figsize=(16, 12))
        
        # Create pivot tables for different metrics
        pivot_time = combined_df.pivot_table(
            index='model_display', 
            columns='prompt', 
            values='duration_sec',
            aggfunc='mean'
        )
        
        pivot_quality = combined_df.pivot_table(
            index='model_display', 
            columns='prompt', 
            values=quality_metric,
            aggfunc='mean'
        )
        
        # Normalize values for color mapping
        norm_time = (pivot_time - pivot_time.min().min()) / (pivot_time.max().max() - pivot_time.min().min())
        norm_quality = (pivot_quality - pivot_quality.min().min()) / (pivot_quality.max().max() - pivot_quality.min().min())
        
        # Create a composite score (low is good for both time and quality)
        composite_score = (norm_time + norm_quality) / 2
        
        # Create a custom color map (green=good, red=bad)
        cmap = LinearSegmentedColormap.from_list("custom_cmap", ["green", "yellow", "red"])
        
        # Plot the heatmap
        plt.subplot(1, 1, 1)
        ax = sns.heatmap(
            composite_score, 
            annot=True, 
            cmap=cmap, 
            fmt='.2f',
            linewidths=0.5
        )
        plt.title('Performance-Quality Trade-off Matrix (Lower is Better)', fontsize=18)
        save_all_formats(plt.gcf(), 'tradeoff_matrix')
        plt.close()
        
        # 2.3 Top Models Comparison
        # Select top 5 models based on composite score
        top_models = composite_score.mean(axis=1).sort_values().index[:5]
        
        # Filter data for top models
        top_models_data = combined_df[combined_df['model_display'].isin(top_models)]
        
        # Create a radar chart for multiple dimensions
        plt.figure(figsize=(12, 10))
        
        # Define metrics to compare (normalize all so lower is better)
        metrics = [
            ('duration_sec', 'Generation Time'),
            (quality_metric, 'Quality Score'),
            ('cpu_percent', 'CPU Usage'),
            ('gpu_mem_start_MB', 'GPU Memory')
        ]
        
        # Number of metrics
        N = len(metrics)
        
        # Create angles for radar chart
        angles = [n / float(N) * 2 * np.pi for n in range(N)]
        angles += angles[:1]  # Close the loop
        
        # Initialize the radar chart
        ax = plt.subplot(111, polar=True)
        
        # Draw one axis per variable and add labels
        plt.xticks(angles[:-1], [label for _, label in metrics])
        
        # Draw y-labels
        ax.set_rlabel_position(0)
        plt.yticks([0.25, 0.5, 0.75], ["0.25", "0.5", "0.75"], color="grey", size=7)
        plt.ylim(0, 1)
        
        # Plot each model
        for model in top_models:
            model_data = combined_df[combined_df['model_display'] == model]
            if len(model_data) == 0:
                continue
                
            # Get average values for the model
            values = []
            for metric, _ in metrics:
                # Normalize metric values (min-max scaling)
                min_val = combined_df[metric].min()
                max_val = combined_df[metric].max()
                avg_val = model_data[metric].mean()
                
                # Scale so lower values are better
                norm_val = (avg_val - min_val) / (max_val - min_val) if max_val > min_val else 0
                values.append(norm_val)
            
            # Close the loop
            values += values[:1]
            
            # Plot the model
            ax.plot(angles, values, linewidth=1, linestyle='solid', label=model)
            ax.fill(angles, values, alpha=0.1)
        
        # Add legend
        plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
        plt.title('Top Models Performance Comparison (Lower is Better)', size=15)
        plt.tight_layout()
        save_all_formats(plt.gcf(), 'top_models_radar')
        plt.close()
        
        # 2.4 Efficiency Ranking Table
        # Create a ranking based on efficiency (balancing quality and speed)
        ranking_df = combined_df.groupby('model_display', group_keys=False).apply(
            lambda x: pd.Series({
                'avg_duration': x['duration_sec'].mean(),
                'avg_quality': x[quality_metric].mean(),
                'efficiency_score': (
                    (x['duration_sec'] - combined_df['duration_sec'].min()) / 
                    (combined_df['duration_sec'].max() - combined_df['duration_sec'].min()) +
                    (x[quality_metric] - combined_df[quality_metric].min()) / 
                    (combined_df[quality_metric].max() - combined_df[quality_metric].min())
                ).mean() / 2
            }), include_groups=False
        ).reset_index()
        
        # Sort by efficiency score (lower is better)
        ranking_df = ranking_df.sort_values('efficiency_score')
        ranking_df['rank'] = range(1, len(ranking_df) + 1)
        
        # Display the ranking table
        print("\n=== Model Efficiency Ranking (Quality vs. Speed) ===")
        print(ranking_df[['rank', 'model_display', 'avg_duration', 'avg_quality', 'efficiency_score']])
        
        # Save the ranking table as a figure
        plt.figure(figsize=(12, len(ranking_df) * 0.4 + 1))
        ax = plt.subplot(111, frame_on=False)
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
        
        table = plt.table(
            cellText=ranking_df[['rank', 'model_display', 'avg_duration', 'avg_quality', 'efficiency_score']].values,
            colLabels=['Rank', 'Model', 'Avg Duration (s)', 'Avg Quality', 'Efficiency Score'],
            cellLoc='center',
            loc='center',
            bbox=[0, 0, 1, 1]
        )
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        table.scale(1, 1.5)
        
        # Highlight top 3 models
        for i in range(3):
            cells = [table[i+1, j] for j in range(5)]
            for cell in cells:
                cell.set_facecolor('#D5F5E3')  # Light green
        
        plt.title('Model Efficiency Ranking (Lower Score is Better)', fontsize=16, y=0.99)
        save_all_formats(plt.gcf(), 'efficiency_ranking')
        plt.close()
    else:
        print("Could not match quality and usage data. Check if model and prompt names match between datasets.")
```

```python
# Prompt 4 Analysis: Advanced Comparisons and Visualizations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.ticker as ticker

# Load your evaluation results
results_df = pd.read_csv('cirrhosis_evaluation_summary.csv')

# Add debug information about the data
print(f"Dataset shape: {results_df.shape}")
print(f"Columns available: {results_df.columns.tolist()}")
print(f"Unique prompts: {results_df['prompt'].unique()}")
print(f"Unique models: {len(results_df['model'].unique())}")

# --------------------- PROMPT 4 SPECIFIC ANALYSIS ---------------------
print("=" * 80)
print("PROMPT 4 ANALYSIS: Exploring Model Performance with Complex Instructions")
print("=" * 80)

# Add the model_display column to the main dataframe first
# Custom mapping: raw Hugging Face names → human-readable names
display_names = {
    "gpt2": "GPT-2-Small",
    "openai-community_gpt2-large": "GPT-2-Large",
    "openai-community_gpt2-medium": "GPT-2-Medium",
    "mistralai_Mistral-7B-Instruct-v0.2": "Mistral 7B",
    "meta-llama_Llama-2-7b-chat-hf": "LLaMA 2 7B",
    "meta-llama_Llama-3.1-8b-Instruct": "LLaMA 3.1 8B",
    "google_gemma-7b-it": "Gemma 7B",
    "01-ai_Yi-6B-Chat": "Yi 6B",
    "HuggingFaceH4_zephyr-7b-beta": "Zephyr 7B",
    "Qwen_Qwen-7B-Chat": "Qwen 7B",
    "Qwen_Qwen2-7B-Instruct": "Qwen2 7B",
    "internlm_internlm2_5-7b-chat": "InternLM2.5 7B",
    "stabilityai_StableBeluga-7B": "StableBeluga 7B",
    "openchat_openchat-3.5-0106": "OpenChat 3.5",
    "NousResearch_Nous-Hermes-2-Yi-34B": "Nous Hermes Yi 34B",
    "mosaicml_mpt-7b-instruct": "MPT 7B",
    "openchat_openchat_3.5": "OpenChat 3.5",
    "TheBloke_openchat_3.5-GPTQ": "OpenChat 3.5 GPTQ"
}

results_df['model_display'] = results_df['model'].map(display_names).fillna(results_df['model'])


# Create filtered dataframes
prompt4_df = results_df[results_df['prompt'] == 'prompt4'].copy()
other_prompts_df = results_df[results_df['prompt'] != 'prompt4'].copy()

print(f"Number of models with prompt4 results: {len(prompt4_df)}")

# Debug gender similarity specifically
if 'categorical_similarity_gender_chi2_statistic' in results_df.columns:
    gender_stats = results_df['categorical_similarity_gender_chi2_statistic'].describe()
    print(f"\nGender Similarity Stats: {gender_stats}")
    print(f"Number of NaN values: {results_df['categorical_similarity_gender_chi2_statistic'].isna().sum()}")
    print(f"Number of zeros: {(results_df['categorical_similarity_gender_chi2_statistic'] == 0).sum()}")
else:
    print("\nGender similarity metric not found in the dataset!")

# 2. Get metrics to analyze - check if they exist first
all_key_metrics = [
    'distribution_metrics_Age_wasserstein',
    'distribution_metrics_Bilirubin_wasserstein',
    'distribution_metrics_Cholesterol_wasserstein',
    'categorical_similarity_Drug_chi2_statistic',
    'feature_correlations_correlation_matrix_distance',
    'medical_consistency_bilirubin_status_relationship_synthetic_D_vs_C_difference'
]

# Filter to only include metrics that actually exist in the dataframe
key_metrics = [metric for metric in all_key_metrics if metric in results_df.columns]
print(f"\nUsing these metrics (found in data): {key_metrics}")

# Create better display names for metrics
metric_display_names = {
    'distribution_metrics_Age_wasserstein': 'Age Distribution Sim.',
    'distribution_metrics_Bilirubin_wasserstein': 'Bilirubin Distribution Sim.',
    'distribution_metrics_Cholesterol_wasserstein': 'Cholesterol Distribution Sim.',
    'categorical_similarity_Drug_chi2_statistic': 'Drug Similarity (Chi2)',
    'feature_correlations_correlation_matrix_distance': 'Correlation Preservation',
    'medical_consistency_bilirubin_status_relationship_synthetic_D_vs_C_difference': 'Bilirubin D vs C Diff (Synth)'
}

# 3. Prompt 4 vs Other Prompts Comparison
# Calculate average metrics for each model across prompt types
prompt_comparison = pd.DataFrame()

for model in results_df['model_display'].unique():
    # Get prompt 4 data
    p4_data = prompt4_df[prompt4_df['model_display'] == model]
    if len(p4_data) == 0:
        continue
        
    # Get other prompts data
    other_data = other_prompts_df[other_prompts_df['model_display'] == model]
    if len(other_data) == 0:
        continue
        
    # For each key metric, calculate the difference (prompt4 - other_prompts_avg)
    model_row = {'model': model}
    
    for metric in key_metrics:
        if metric in p4_data.columns and metric in other_data.columns:
            # Check if we have non-NaN values
            p4_vals = p4_data[metric].dropna()
            other_vals = other_data[metric].dropna()
            
            if len(p4_vals) == 0 or len(other_vals) == 0:
                continue
                
            # For metrics where lower is better
            if 'wasserstein' in metric or 'chi2' in metric or 'correlation_matrix_distance' in metric:
                p4_val = p4_vals.mean()
                other_val = other_vals.mean()
                # Negative means prompt4 is better (lower)
                diff = p4_val - other_val
                # Normalize by other_val for percentage difference
                if other_val != 0:
                    pct_diff = (diff / other_val) * 100
                else:
                    pct_diff = 0
                
                model_row[f"{metric}_diff"] = diff
                model_row[f"{metric}_pct"] = pct_diff
            else:
                # For metrics where higher is better
                p4_val = p4_vals.mean()
                other_val = other_vals.mean()
                # Positive means prompt4 is better (higher)
                diff = p4_val - other_val
                # Normalize by other_val for percentage difference
                if other_val != 0:
                    pct_diff = (diff / other_val) * 100
                else:
                    pct_diff = 0
                    
                model_row[f"{metric}_diff"] = diff
                model_row[f"{metric}_pct"] = pct_diff
                
    prompt_comparison = pd.concat([prompt_comparison, pd.DataFrame([model_row])])

# Display the prompt comparison results
print("\nPrompt 4 vs Other Prompts Performance Difference (Percentage):")
comparison_display = pd.DataFrame()
comparison_display['Model'] = prompt_comparison['model']

for metric in key_metrics:
    if f"{metric}_pct" in prompt_comparison.columns:
        comparison_display[metric_display_names[metric]] = prompt_comparison[f"{metric}_pct"].round(2)

print(comparison_display)

# 4. Visualize the prompt 4 vs other prompts comparison
plt.figure(figsize=(15, 10))
comparison_melted = pd.melt(
    comparison_display, 
    id_vars=['Model'], 
    var_name='Metric', 
    value_name='Percentage Difference'
)

# Create a diverging colormap for better visualization
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Create heatmap
plt.figure(figsize=(14, len(comparison_display) * 0.8))
heatmap_data = comparison_display.set_index('Model')
sns.heatmap(
    heatmap_data, 
    annot=True, 
    cmap=cmap, 
    center=0, 
    fmt='.1f',
    linewidths=0.5,
    vmin=-50,  # Limit to +/- 50% for better color scale
    vmax=50
)
plt.title('Prompt 4 Performance Difference vs Other Prompts (%)', fontsize=16)
save_all_formats(plt.gcf(), 'prompt4_comparison_heatmap')
plt.show()

# 5. Model Ranking on Prompt 4
# Create a normalized score for each model on prompt 4
normalized_scores = pd.DataFrame()
normalized_scores['model'] = prompt4_df['model_display'].unique()

# Normalize each metric (0-1 scale where 1 is best)
for metric in key_metrics:
    if metric in prompt4_df.columns:
        metric_vals = []
        for model in normalized_scores['model']:
            model_data = prompt4_df[prompt4_df['model_display'] == model]
            if metric in model_data.columns and not model_data[metric].isna().all():
                metric_vals.append(model_data[metric].mean())
            else:
                metric_vals.append(np.nan)
                
        normalized_scores[metric] = metric_vals
        
        # Skip if all values are NaN
        if normalized_scores[metric].isna().all():
            continue
            
        # For metrics where lower is better
        if 'wasserstein' in metric or 'chi2' in metric or 'correlation_matrix_distance' in metric:
            min_val = normalized_scores[metric].min(skipna=True)
            max_val = normalized_scores[metric].max(skipna=True)
            if max_val > min_val:
                normalized_scores[metric] = 1 - ((normalized_scores[metric] - min_val) / (max_val - min_val))
            else:
                normalized_scores[metric] = normalized_scores[metric].apply(lambda x: 1.0 if not pd.isna(x) else x)
        else:
            # For metrics where higher is better
            min_val = normalized_scores[metric].min(skipna=True)
            max_val = normalized_scores[metric].max(skipna=True)
            if max_val > min_val:
                normalized_scores[metric] = (normalized_scores[metric] - min_val) / (max_val - min_val)
            else:
                normalized_scores[metric] = normalized_scores[metric].apply(lambda x: 1.0 if not pd.isna(x) else x)

# Only use metrics that actually have data
valid_metrics = [col for col in key_metrics if col in normalized_scores.columns and not normalized_scores[col].isna().all()]
print(f"\nValid metrics for ranking: {valid_metrics}")

# Calculate average normalized score using only available metrics
if valid_metrics:
    normalized_scores['overall_score'] = normalized_scores[valid_metrics].mean(axis=1)
else:
    normalized_scores['overall_score'] = 0.5  # Default if no metrics available

# Rank models
ranked_models = normalized_scores.sort_values('overall_score', ascending=False)

print("\nModel Ranking on Prompt 4 (Higher is Better):")
for i, (_, row) in enumerate(ranked_models.iterrows()):
    print(f"{i+1}. {row['model']}: {row['overall_score']:.4f}")

# 6. Top Models Performance Radar Chart for Prompt 4
if len(valid_metrics) >= 3:  # Only create radar chart if we have enough metrics
    plt.figure(figsize=(12, 10))
    top_models = ranked_models.head(5)['model'].values
    
    # Prepare data for radar chart
    categories = [metric_display_names.get(m, m) for m in valid_metrics]
    N = len(categories)
    
    # Create angles for radar chart
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]  # Close the loop
    
    # Initialize the radar chart
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
    
    # Add each top model to the chart
    for model in top_models:
        model_data = normalized_scores[normalized_scores['model'] == model]
        if len(model_data) > 0:
            values = model_data[valid_metrics].values[0].tolist()
            values += values[:1]  # Close the loop
            
            ax.plot(angles, values, linewidth=2, label=model)
            ax.fill(angles, values, alpha=0.1)
    
    # Fix radar chart
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    plt.xticks(angles[:-1], categories)
    ax.set_ylim(0, 1)
    
    # Draw axis lines for better readability
    ax.set_rlabel_position(0)
    plt.yticks([0.25, 0.5, 0.75], ["0.25", "0.5", "0.75"], color="grey", size=8)
    
    # Add legend
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.title("Top 5 Models Performance on Prompt 4", size=15)
    save_all_formats(plt.gcf(), 'prompt4_top_models_radar')
    plt.show()

# 7. Feature Importance for Prompt 4 Success
# Which metrics most strongly correlate with overall success on prompt 4?
correlation_with_overall = {}

for metric in valid_metrics:
    if metric in normalized_scores.columns and 'overall_score' in normalized_scores.columns:
        corr = normalized_scores[metric].corr(normalized_scores['overall_score'])
        correlation_with_overall[metric_display_names.get(metric, metric)] = corr

# Sort by absolute correlation
correlation_df = pd.DataFrame({
    'Metric': list(correlation_with_overall.keys()),
    'Correlation with Overall Score': list(correlation_with_overall.values())
}).sort_values('Correlation with Overall Score', ascending=False)

print("\nMetric Importance for Prompt 4 Success:")
print(correlation_df)

# Visualize metric importance
if len(correlation_df) > 0:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Correlation with Overall Score', y='Metric', data=correlation_df)
    plt.title('Metric Correlation with Overall Prompt 4 Performance', fontsize=14)
    plt.xlabel('Correlation Coefficient', fontsize=12)
    plt.ylabel('Metric', fontsize=12)
    save_all_formats(plt.gcf(), 'prompt4_metric_importance')
    plt.show()

# 8. Alternative Prompt Difficulty Analysis using basic metrics
print("\nAlternative Prompt Difficulty Analysis...")

# Try with metrics we're more likely to have
metrics_to_try = [
    'numerical_statistics_Age_mean_difference',
    'numerical_statistics_Bilirubin_mean_difference',
    'distribution_metrics_Age_wasserstein',
    'feature_correlations_correlation_matrix_distance'
]

# Check which ones exist
alt_metrics = [m for m in metrics_to_try if m in results_df.columns]
print(f"Using these metrics for difficulty: {alt_metrics}")

if alt_metrics:
    alt_difficulty = pd.DataFrame()
    
    for prompt in results_df['prompt'].unique():
        prompt_data = results_df[results_df['prompt'] == prompt]
        prompt_row = {'prompt': prompt}
        
        for metric in alt_metrics:
            if not prompt_data[metric].isna().all():
                # For all these metrics, lower is better
                prompt_row[metric] = prompt_data[metric].mean()
        
        alt_difficulty = pd.concat([alt_difficulty, pd.DataFrame([prompt_row])])
    
    # Calculate difficulty score (lower metric values = better performance)
    for metric in alt_metrics:
        if metric in alt_difficulty.columns:
            min_val = alt_difficulty[metric].min()
            max_val = alt_difficulty[metric].max()
            if max_val > min_val:
                # Convert to 0-1 scale where 1 is best (lowest raw value)
                alt_difficulty[f"{metric}_norm"] = 1 - ((alt_difficulty[metric] - min_val) / (max_val - min_val))
    
    # Get normalized metrics
    norm_cols = [col for col in alt_difficulty.columns if col.endswith('_norm')]
    
    if norm_cols:
        # Higher difficulty score = easier prompt
        alt_difficulty['difficulty_score'] = alt_difficulty[norm_cols].mean(axis=1)
        
        # Create prompt ranking (ascending = hardest first)
        prompt_ranking = alt_difficulty.sort_values('difficulty_score', ascending=True)
        
        print("\nPrompt Difficulty Ranking (Lower Score = More Difficult):")
        for i, (_, row) in enumerate(prompt_ranking.iterrows()):
            print(f"{i+1}. {row['prompt']}: {row['difficulty_score']:.4f}")
        
        # 9. Visualize prompt difficulty
        plt.figure(figsize=(12, 6))
        
        # Add color column
        prompt_ranking['is_prompt4'] = prompt_ranking['prompt'] == 'prompt4'
        
        # Create bar plot with hue
        ax = sns.barplot(x='prompt', y='difficulty_score', hue='is_prompt4', 
                         data=prompt_ranking, palette={True: 'red', False: 'blue'}, legend=False)
        
        plt.title('Prompt Difficulty (Lower Score = More Difficult)', fontsize=14)
        plt.xlabel('Prompt', fontsize=12)
        plt.ylabel('Average Performance Score', fontsize=12)
        plt.ylim(0, 1)
        
        for i, v in enumerate(prompt_ranking['difficulty_score']):
            plt.text(i, v + 0.02, f"{v:.3f}", ha='center')
        
        save_all_formats(plt.gcf(), 'prompt_difficulty')
        plt.show()
    else:
        print("No normalized metrics available for prompt difficulty ranking")
else:
    print("No metrics found for difficulty calculation")

# 10. Create a comprehensive prompt 4 performance summary table
summary_table = pd.DataFrame({
    'Model': ranked_models['model'],
    'Overall Score': ranked_models['overall_score'].round(3),
})

# Add important metrics if they exist
if len(correlation_df) >= 3:
    top_metrics = correlation_df.head(3)['Metric'].values
    for metric_name in top_metrics:
        for orig_name, display_name in metric_display_names.items():
            if display_name == metric_name and orig_name in ranked_models.columns:
                summary_table[metric_name] = ranked_models[orig_name].round(3)

print("\nPrompt 4 Performance Summary (Top Models):")
print(summary_table.head(10))

# Save the prompt 4 analysis results
summary_table.to_csv('prompt4_analysis_summary.csv', index=False)
print("\nPrompt 4 analysis completed and saved to CSV!")
```

```python
# Extract and Analyze Enhanced Privacy Metrics - Fixed Version
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.ticker as ticker
import re

# Load evaluation results
results_df = pd.read_csv('cirrhosis_evaluation_summary.csv')

print("=" * 80)
print("ENHANCED PRIVACY AND QUALITY TRADE-OFF ANALYSIS")
print("=" * 80)

# First, let's see which privacy metrics are available
privacy_columns = [col for col in results_df.columns if 
                  'privacy' in col.lower() or 
                  'k_anonymity' in col.lower() or 
                  'identifiability' in col.lower() or 
                  'attack' in col.lower() or
                  'disclosure' in col.lower() or
                  'nn_distance' in col.lower()]

print(f"\nAvailable privacy metrics ({len(privacy_columns)}):")
for col in privacy_columns[:10]:  # Show first 10 for brevity
    print(f"- {col}")
if len(privacy_columns) > 10:
    print(f"...and {len(privacy_columns) - 10} more")

# Define quality metrics for comparison
quality_columns = [
    'distribution_metrics_age_wasserstein',
    'distribution_metrics_bmi_wasserstein',
    'feature_correlations_correlation_matrix_distance',
    'new_metrics_TSTR_accuracy',
    'medical_consistency_hba1c_diabetes_relationship_synthetic_difference'
]

# Check which ones actually exist in the data
quality_cols_available = [col for col in quality_columns if col in results_df.columns]

print(f"\nAvailable quality metrics ({len(quality_cols_available)}):")
for col in quality_cols_available:
    print(f"- {col}")

# Helper function to safely calculate mean, min, max values
def safe_stats(series):
    """Calculate stats safely, even with string values."""
    # Try to convert to numeric first
    numeric_series = pd.to_numeric(series, errors='coerce')
    # Calculate stats only on non-NaN values
    values = numeric_series.dropna()
    
    if len(values) == 0:
        return None, None, None
    
    return values.mean(), values.min(), values.max()

# 1. Create a summary of privacy metrics by prompt
print("\n1. PRIVACY METRICS BY PROMPT")
print("-" * 40)

# Privacy metrics summary for each prompt
privacy_summary = []

# Look for the most important privacy metrics
key_privacy_metrics = [
    'privacy_assessment_age_gender_violations_violation_rate',  # K-anonymity violation rate
    'advanced_metrics_privacy_metrics_nn_distance_ratio',      # NN distance ratio
    'advanced_metrics_privacy_metrics_identifiability_score',  # Identifiability score
    'privacy_assessment_overall_privacy_score'                # Overall privacy score
]

# Check which ones actually exist
key_privacy_available = [col for col in key_privacy_metrics if col in results_df.columns]

for prompt in results_df['prompt'].unique():
    prompt_data = results_df[results_df['prompt'] == prompt]
    
    # Basic prompt info
    prompt_stats = {'prompt': prompt, 'num_models': len(prompt_data)}
    
    # Add statistics for each privacy metric
    for metric in privacy_columns:
        if metric in prompt_data.columns:
            # Try to get numeric values safely
            mean_val, min_val, max_val = safe_stats(prompt_data[metric])
            
            if mean_val is not None:
                prompt_stats[f"{metric}_mean"] = mean_val
                prompt_stats[f"{metric}_min"] = min_val
                prompt_stats[f"{metric}_max"] = max_val
    
    privacy_summary.append(prompt_stats)

# Convert to DataFrame
privacy_df = pd.DataFrame(privacy_summary)

# Create a more readable summary table
if not privacy_df.empty:
    display_cols = ['prompt', 'num_models']
    for metric in key_privacy_available:
        mean_col = f"{metric}_mean"
        if mean_col in privacy_df.columns:
            # Create readable name
            if 'violation_rate' in metric:
                display_name = 'K-Anonymity Violation Rate'
                # Lower is better
                best_idx = privacy_df[mean_col].fillna(float('inf')).idxmin()
                better_dir = 'lower'
            elif 'nn_distance_ratio' in metric:
                display_name = 'NN Distance Ratio'
                # Higher is better
                best_idx = privacy_df[mean_col].fillna(float('-inf')).idxmax()
                better_dir = 'higher'
            elif 'identifiability_score' in metric:
                display_name = 'Identifiability Score'
                # Lower is better
                best_idx = privacy_df[mean_col].fillna(float('inf')).idxmin()
                better_dir = 'lower'
            elif 'overall_privacy_score' in metric:
                display_name = 'Overall Privacy Score'
                # Higher is better
                best_idx = privacy_df[mean_col].fillna(float('-inf')).idxmax()
                better_dir = 'higher'
            else:
                display_name = metric.split('_')[-1]
                # Default: lower is better
                best_idx = privacy_df[mean_col].fillna(float('inf')).idxmin()
                better_dir = 'lower'
            
            print(f"\n{display_name} by prompt (lower {'is better' if better_dir == 'lower' else 'is worse'}):")
            for i, row in privacy_df.iterrows():
                prompt = row['prompt']
                if mean_col in row and not pd.isna(row[mean_col]):
                    value = row[mean_col]
                    best = "★" if i == best_idx else ""
                    print(f"  {prompt}: {value:.4f} {best}")

# 2. Quality metrics summary by prompt
print("\n2. QUALITY METRICS BY PROMPT")
print("-" * 40)

quality_summary = []

for prompt in results_df['prompt'].unique():
    prompt_data = results_df[results_df['prompt'] == prompt]
    
    # Basic prompt info
    prompt_stats = {'prompt': prompt, 'num_models': len(prompt_data)}
    
    # Add statistics for each quality metric
    for metric in quality_cols_available:
        if metric in prompt_data.columns:
            # Use safe stats function
            mean_val, min_val, max_val = safe_stats(prompt_data[metric])
            
            if mean_val is not None:
                prompt_stats[f"{metric}_mean"] = mean_val
                prompt_stats[f"{metric}_min"] = min_val
                prompt_stats[f"{metric}_max"] = max_val
    
    quality_summary.append(prompt_stats)

# Convert to DataFrame
quality_df = pd.DataFrame(quality_summary)

# Display quality metrics
if not quality_df.empty:
    for metric in quality_cols_available:
        mean_col = f"{metric}_mean"
        if mean_col in quality_df.columns:
            # Create readable name
            if 'wasserstein' in metric:
                feature = metric.split('_')[2]
                display_name = f"{feature.capitalize()} Distribution Similarity"
                # Lower is better
                best_idx = quality_df[mean_col].fillna(float('inf')).idxmin()
                better_dir = 'lower'
            elif 'correlation_matrix_distance' in metric:
                display_name = 'Correlation Preservation'
                # Lower is better
                best_idx = quality_df[mean_col].fillna(float('inf')).idxmin()
                better_dir = 'lower'
            elif 'TSTR_accuracy' in metric:
                display_name = 'Train-Synthetic Test-Real Accuracy'
                # Higher is better
                best_idx = quality_df[mean_col].fillna(float('-inf')).idxmax()
                better_dir = 'higher'
            elif 'synthetic_difference' in metric:
                display_name = 'HbA1c-Diabetes Relationship'
                # Depends on context, assuming lower is better for difference
                best_idx = quality_df[mean_col].fillna(float('inf')).idxmin()
                better_dir = 'lower'
            else:
                display_name = metric
                # Default: lower is better
                best_idx = quality_df[mean_col].fillna(float('inf')).idxmin()
                better_dir = 'lower'
            
            print(f"\n{display_name} by prompt (lower {'is better' if better_dir == 'lower' else 'is worse'}):")
            for i, row in quality_df.iterrows():
                prompt = row['prompt']
                if mean_col in row and not pd.isna(row[mean_col]):
                    value = row[mean_col]
                    best = "★" if i == best_idx else ""
                    print(f"  {prompt}: {value:.4f} {best}")

# 3. Privacy-Quality Trade-off Analysis
print("\n3. PRIVACY-QUALITY TRADE-OFF ANALYSIS")
print("-" * 40)

# Simple summary comparing prompt performance
print("\nPrompt summary (ranking by metric):")
print("=" * 60)
print(f"{'Prompt':<10} | {'Privacy':<20} | {'Quality':<20}")
print("-" * 60)

# Let's pick the most reliable metrics 
privacy_metric = 'advanced_metrics_privacy_metrics_nn_distance_ratio_mean'
quality_metric = 'distribution_metrics_age_wasserstein_mean'

# Check if these metrics exist
privacy_available = privacy_metric in privacy_df.columns
quality_available = quality_metric in quality_df.columns

if privacy_available and quality_available:
    # Get rankings for each prompt
    privacy_ranks = privacy_df.sort_values(privacy_metric, ascending=False)['prompt'].tolist()
    quality_ranks = quality_df.sort_values(quality_metric, ascending=True)['prompt'].tolist()  # Lower is better
    
    for prompt in results_df['prompt'].unique():
        privacy_rank = privacy_ranks.index(prompt) + 1 if prompt in privacy_ranks else "N/A"
        quality_rank = quality_ranks.index(prompt) + 1 if prompt in quality_ranks else "N/A"
        print(f"{prompt:<10} | #{privacy_rank:<19} | #{quality_rank:<19}")
    
    # Create a simple trade-off metric
    tradeoff_data = []
    for prompt in results_df['prompt'].unique():
        prompt_row = {'prompt': prompt}
        
        # Get normalized scores (0-1, higher is better)
        if prompt in privacy_ranks:
            # Convert rank to score (inverse, so 1st = 1.0, last = 0.0)
            prompt_row['privacy_score'] = 1.0 - ((privacy_ranks.index(prompt)) / max(1, len(privacy_ranks)-1))
        else:
            prompt_row['privacy_score'] = 0.0
            
        if prompt in quality_ranks:
            # Convert rank to score (inverse, so 1st = 1.0, last = 0.0)
            prompt_row['quality_score'] = 1.0 - ((quality_ranks.index(prompt)) / max(1, len(quality_ranks)-1))
        else:
            prompt_row['quality_score'] = 0.0
            
        # Calculate combined score
        prompt_row['combined_score'] = (prompt_row['privacy_score'] + prompt_row['quality_score']) / 2
        
        tradeoff_data.append(prompt_row)
    
    # Create DataFrame and sort by combined score
    tradeoff_df = pd.DataFrame(tradeoff_data).sort_values('combined_score', ascending=False)
    
    print("\nPrompt ranking by combined privacy-quality score (higher is better):")
    for i, (_, row) in enumerate(tradeoff_df.iterrows()):
        print(f"{i+1}. {row['prompt']}: {row['combined_score']:.4f} " +
              f"(Privacy: {row['privacy_score']:.2f}, Quality: {row['quality_score']:.2f})")
    
    # Create a simple scatter plot
    plt.figure(figsize=(10, 8))
    
    # Plot each prompt
    for _, row in tradeoff_df.iterrows():
        plt.scatter(row['privacy_score'], row['quality_score'], s=100)
        plt.text(row['privacy_score']+0.02, row['quality_score'], row['prompt'], fontsize=12)
    
    # Add diagonal line and labels
    plt.plot([0, 1], [0, 1], 'k--', alpha=0.3)
    plt.xlabel('Privacy Score (higher is better)', fontsize=12)
    plt.ylabel('Quality Score (higher is better)', fontsize=12)
    plt.title('Privacy vs Quality Trade-off by Prompt', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    
    save_all_formats(plt.gcf(), 'privacy_quality_tradeoff')
    plt.show()

# 4. Model-Level Analysis for Prompt 4
print("\n4. MODEL-LEVEL ANALYSIS FOR PROMPT 4")
print("-" * 40)

prompt4_df = results_df[results_df['prompt'] == 'prompt4'].copy()

if len(prompt4_df) > 0:
    # Add a display name column

    # Custom mapping: raw Hugging Face names → human-readable names
    display_names = {
        "gpt2": "GPT-2-Small",
        "openai-community_gpt2-medium": "GPT-2-Medium",
        "openai-community_gpt2-large": "GPT-2-Large",
        "mistralai_Mistral-7B-Instruct-v0.2": "Mistral 7B",
        "meta-llama_Llama-2-7b-chat-hf": "LLaMA 2 7B",
        "meta-llama_Llama-3.1-8b-Instruct": "LLaMA 3.1 8B",
        "meta-llama_Llama-2-13b-hf": "LLaMA 2 13B",
        "meta-llama_Llama-2-13b-chat-hf": "LLaMA 2 13B Chat",
        "meta-llama_Meta-Llama-3-8B": "LLaMA 3 8B",
        "google_gemma-7b-it": "Gemma 7B",
        "01-ai_Yi-6B-Chat": "Yi 6B",
        "HuggingFaceH4_zephyr-7b-beta": "Zephyr 7B",
        "Qwen_Qwen2-7B-Instruct": "Qwen2 7B",
        "stabilityai_StableBeluga-7B": "StableBeluga 7B",
        "openchat_openchat-3.5-0106": "OpenChat 3.5-0106",
        "openchat_openchat_3.5": "OpenChat 3.5",
        "TheBloke_openchat_3.5-GPTQ": "OpenChat 3.5 GPTQ",
        "NousResearch_Nous-Hermes-2-Yi-34B": "Nous Hermes Yi 34B",
        "NousResearch_Nous-Hermes-2-Mistral-7B-DPO": "Nous Hermes Mistral 7B DPO",
        "mosaicml_mpt-7b-instruct": "Mosaic MPT 7B",
        "internlm_internlm2_5-7b-chat": "InternLM2.5 7B",
        "lmsys_vicuna-13b-v1.5": "Vicuna 13B",
        "Expert68_llama2_13b_instructed_version2": "LLaMA 2 13B Instruct (Exp68)",
        "meta-llama_Llama-4-Maverick-17B-128E-Instruct": "LLaMA 4 Maverick 17B"
    }
    
    prompt4_df['model_display'] = prompt4_df['model'].map(display_names).fillna(prompt4_df['model'])
    
    # Check for privacy and quality metrics at model level
    privacy_model_metric = 'advanced_metrics_privacy_metrics_nn_distance_ratio'
    quality_model_metric = 'distribution_metrics_Age_wasserstein'
    
    privacy_available = privacy_model_metric in prompt4_df.columns
    quality_available = quality_model_metric in prompt4_df.columns
    
    if privacy_available and quality_available:
        # Convert to numeric, handling errors
        prompt4_df[privacy_model_metric] = pd.to_numeric(prompt4_df[privacy_model_metric], errors='coerce')
        prompt4_df[quality_model_metric] = pd.to_numeric(prompt4_df[quality_model_metric], errors='coerce')
        
        # Drop rows with missing values
        valid_models = prompt4_df.dropna(subset=[privacy_model_metric, quality_model_metric])
        
        if len(valid_models) > 0:
            # Create normalized scores (higher is better)
            privacy_min = valid_models[privacy_model_metric].min()
            privacy_max = valid_models[privacy_model_metric].max()
            privacy_range = max(privacy_max - privacy_min, 0.0001)  # Avoid division by zero
            
            quality_min = valid_models[quality_model_metric].min()
            quality_max = valid_models[quality_model_metric].max()
            quality_range = max(quality_max - quality_min, 0.0001)  # Avoid division by zero
            
            # For privacy (NN distance ratio), higher is better
            valid_models['privacy_score'] = (valid_models[privacy_model_metric] - privacy_min) / privacy_range
            
            # For quality (wasserstein), lower is better
            valid_models['quality_score'] = 1 - (valid_models[quality_model_metric] - quality_min) / quality_range
            
            # Calculate combined score
            valid_models['combined_score'] = (valid_models['privacy_score'] + valid_models['quality_score']) / 2
            
            # Sort and display top models
            top_models = valid_models.sort_values('combined_score', ascending=False)
            
            print("\nTop models by combined privacy-quality score (Prompt 4):")
            for i, (_, row) in enumerate(top_models.head(5).iterrows()):
                print(f"{i+1}. {row['model_display']}: {row['combined_score']:.4f} " +
                      f"(Privacy: {row['privacy_score']:.2f}, Quality: {row['quality_score']:.2f})")
            
            # Create a scatter plot
            plt.figure(figsize=(12, 10))
            
            # Plot each model
            for _, row in valid_models.iterrows():
                plt.scatter(row['privacy_score'], row['quality_score'], s=100)
                plt.text(row['privacy_score']+0.02, row['quality_score']+0.01, 
                        row['model_display'], fontsize=10)
            
            # Add labels and lines
            plt.plot([0, 1], [0, 1], 'k--', alpha=0.3)
            plt.xlabel('Privacy Score (higher is better)', fontsize=12)
            plt.ylabel('Quality Score (higher is better)', fontsize=12)
            plt.title('Privacy vs Quality Trade-off by Model (Prompt 4)', fontsize=14)
            plt.grid(True, alpha=0.3)
            plt.xlim(-0.05, 1.05)
            plt.ylim(-0.05, 1.05)
            
            save_all_formats(plt.gcf(), 'model_privacy_quality_tradeoff')
            plt.show()
```

```python
import numpy as np

def plot_model_radar(model_name):
    prompts = ['prompt1', 'prompt2', 'prompt3', 'prompt4']
    scores = harmonic_table.loc[model_name, prompts].fillna(0).values
    angles = np.linspace(0, 2 * np.pi, len(prompts), endpoint=False).tolist()
    scores = np.concatenate((scores, [scores[0]]))  # close the circle
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.plot(angles, scores, 'o-', linewidth=2, label=model_name)
    ax.fill(angles, scores, alpha=0.25)
    ax.set_thetagrids(np.degrees(angles[:-1]), prompts)
    ax.set_ylim(0, 1.1)
    ax.set_title(f"Harmonic Score Radar Plot – {model_name}", fontsize=14)
    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
    save_all_formats(plt.gcf(), f'{model_name}_radar_plot')
    plt.show()

# Example: plot for a top model
plot_model_radar("Nous Hermes Yi 34B")
```
